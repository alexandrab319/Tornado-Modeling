```python
import math 
from mpl_toolkits.mplot3d import axes3d 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### The Field is not conservative, this shows the direction in which the wind moves

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

x, y, z = np.meshgrid(np.arange(-3, 3, 1), 
                      np.arange(-3, 3, 1), 
                      np.arange(0, 3, 1))

u = (-1 * x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
v = (x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
w = 2**0.5 / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5

ax.quiver(x, y, z, u, v, w, color = 'black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

### Event horizon of the tornado

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

x, y = np.meshgrid(np.arange(-6, 6, 1), 
                    np.arange(-6, 6, 1))
                    
u = x
v = y
w = np.log(x**2 + y**2)

ax.plot_surface(u, v, w, color = 'black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

ax.view_init(60, 45)
plt.subplot();
```

### The movement and speed of the wind around the event horizon of the tornado

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

x, y = np.meshgrid(np.arange(-6, 6, 1), 
                    np.arange(-6, 6, 1))
                    
u = x
v = y
w = np.log(x**2 + y**2)

ax.plot_surface(u, v, w, color = 'royalblue', alpha = 0.5)

x, y, z = np.meshgrid(np.arange(-3, 3, 1), 
                      np.arange(-3, 3, 1), 
                      np.arange(0, 3, 1))

s = 10 
u = s * (-1 * x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
v = s * (x - y) / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5
w = s * 2**0.5 / ((-1 * x - y)**2 + (x - y)**2 + 2)**0.5


for j in range(len(x)):
    for i in range(len(x[j])):
        for k in range(len(x[i,j])):
            if x[j][i][k]**2 + y[j][i][k]**2 >=1:
                ax.quiver(x[j][i][k], y[j][i][k], z[j][i][k], u[j][i][k], v[j][i][k], w[j][i][k], length=0.2, color='black')


plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")
ax.view_init(45, 45)
plt.subplot();
```

### The trajectory of the object the wind is applying to

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

t = np.linspace(0, 2 * math.pi, 100)
t = np.array(t)

r = 1
x = r * np.cos(t)
y = r * np.sin(t)
z = t

ax.plot(x, y, z, color = 'black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

### This plane illustrates the impact of the radius of the trajectory and the size of the force vector on the work performed on the object

```python
# (sqrt(2) + r^2) * r * e^-r * 2 * (pi)^2

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

k, r = np.meshgrid(np.arange(-3, 3, 0.2), np.arange(0, 3, 0.2))

u = k
v = r
w = k * ((2**0.5 + r**2) * r * np.e**(-r) * 2 * (np.pi)**2)

ax.plot_surface(u, v, w, color = 'black')

plt.xlabel("k")
plt.ylabel("$r_0$")
ax.set_zlabel("Z")

plt.subplot();
```

### The starting position is rotated counterclockwise to a different quadrant, and the path becomes clockwise instead of counterclockwise as it originally was

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

t = np.linspace(0, 2 * math.pi, 100)
t = np.array(t)

r = 1
x = r * np.sin(t)
y = r * np.cos(t)
z = t

ax.plot(x, y, z, color = 'black')

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

### The force appears to have a greater influence when the radius is not changed, especially when the radius is zero. As the radius increases, the force appears to have a weaker impact compared to the original trajectory

```python
# (sqrt(2) + r^2) * r * e^-r * 2 * (pi)^2

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

k, r = np.meshgrid(np.arange(-3, 3, 0.2), np.arange(0, 3, 0.2))

u = k
v = r
w = k * ((2**1.5) * r * np.e**(-r) * 2 * (np.pi)**2)

ax.plot_surface(u, v, w, color = 'black', alpha = 0.8)

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

### Parametrization of the cylinder

```python
# (sqrt(2) + r^2) * r * e^-r * 2 * (pi)^2

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection = '3d')

t, r = np.meshgrid(np.linspace(0,2 * np.pi, 100), np.linspace(0, 2, 100))

x = r * np.cos(t)
y = r * np.sin(t)
z = t

ax.plot_surface(x, y, z)

plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")

plt.subplot();
```

### The inward flux rallies as you move farther away from the tornado, quickly increasing between a radius of 2.5 and 5.0, before rapidly decreasing again

```python
fig = plt.figure(figsize = (10,10))

r = np.linspace(0, 20, 20)
y = np.pi**3 * r**3 * np.e**(-r)

plt.plot(r, y)
plt.xlabel("$r_0$")
plt.ylabel("Inward Flux")

plt.subplot();
```

### The upward air flux increases at a very low height, then decreases linearly after reaching a height of about 2

```python
fig = plt.figure(figsize = (10,10))

z = np.linspace(0, 10, 25)
y = 2 * 2**0.5 * np.pi * (2 * np.pi - z) * (2 * (-np.e**(z - np.e**z) - np.e**(-np.e**z)) + 2 - np.e**(2*z - np.e**z))

plt.plot(z, y)
plt.xlabel("Height")
plt.ylabel("Upwards Air Flux")

plt.subplot();
```
